```python
# -*- encoding: utf-8 -*-
import jieba

# 词频
del_list=[]
word_list = []
with open('./tinglist.txt', 'r', encoding='utf-8') as f1:
    for word in f1.readlines():
       del_list.append(word)
del_list = list(set(del_list))

# f=open('./tinglist.txt', 'r')
# for i in f.readlines(): 
#     del_list.append(i.decode('utf-8'))
with open('./fuqindeshenfenyangben 1.txt', 'r', encoding='utf-8') as f2:
    for sentence in f2.readlines():
        seg_list = jieba.cut(sentence.strip(), cut_all = False)
        for i in seg_list:
            if i not in del_list and i != ' ':
                word_list.append(i)
                
for word in word_list:
    print(word)
```
