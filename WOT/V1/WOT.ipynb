```python
import os
import time
import graphlab as gl

gl.canvas.set_target('ipynb')
# reset
# gl.canvas.set_target('browser')
```

```python
data_folder = './data/'
source = os.listdir(data_folder)
for idx, filename in enumerate(source):
    print('fileID: %s, fileName: %s, fileSize: %s M'%(idx, filename, os.path.getsize(data_folder + filename) / float(1024 ** 2)))
```

## A glimpse to `table_1.tsv` 

>battles on a particular day

```python
%%time
# if using Pandas, importing data would be a mission impossible
battle = gl.SFrame.read_csv(data_folder + source[2], 
                            header=False, 
                            delimiter='\t')
```

```python
battle
```

> `table_1.tsv` contains 43,043,874 rows * 36 columns 

### How many users?

> 279,313 users

```python
# X2 = user identifier (SPA ID)
battle['X2'] = battle['X2'].astype(str)
battle['X2'].sketch_summary()
```

```python
%%time
# double check
print(len(set(list(battle['X2']))))
```

### How many battles?

> 1,825,553 battles

```python
# X8 = battle identifier
battle['X8'] = battle['X8'].astype(str)
battle['X8'].sketch_summary()
```

```python
%%time
# double check
print(len(set(list(battle['X8']))))
```

### Amount of time user was in the battle total

```python
# X12 = amount of time user was in the battle total
battle['X12'].show(view='Numeric')
```

### Battle start time

> earliest start time: `2018-03-01 00:00:00`

```python
%%time
# X30 = battle start time
battle['stTimestamp'] = battle['X30'].apply(lambda x: time.mktime(time.strptime(x, "%Y-%m-%d %H:%M:%S")))
```

```python
battle['stTimestamp'].sketch_summary()
```

```python
# calculate the exact start time
print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(1519891200.0)))
```

```python
battle['stInterval'] = battle['stTimestamp'] - 1519891200.0
battle['stInterval'].show(view='Numeric')
```
