```python
data_1.rename({4: 'date'}, axis=1, inplace=True)
data_1.head()
```

```python
pattern_date = re.compile(r'(\d+)-(\d+)-(\d+)')

def get_date(string):
    result = ''
    if pattern_date.search(str(string)) != None:
        raw_result = pattern_date.search(str(string))
        result = raw_result.group(1) + '-' + raw_result.group(2) + '-' + raw_result.group(3)
    return result
```

```python
data_1['new_date'] = data_1['date'].apply(get_date)
data_1.head()
```

```python
training_data = pd.read_excel('./0314/machineLearning/new_Annotated_3_type_cllasification.xlsx', header=0)
training_data.head()
```

```python
set(list(training_data['Issue1']))
```

```python
training_data = training_data[training_data['Issue1'].apply(lambda x: str(x) != 'nan')]
training_data.shape
```

```python
training_data.rename(columns={'Issue1': 'target'}, inplace=True)
```

```python
pattern_url = re.compile(r'h?ttp[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]| (?:%[0-9a-fA-F][0-9a-fA-F]))+')
pattern_blank = re.compile(r'\s+')
pattern_special_char = re.compile(r' | |\[|\]|\/| | ')
pattern_number = re.compile(r'\d+')
pattern_allChar = re.compile(r'(^[a-z]+[a-z]$)')
pattern_allChar2 = re.compile(r'(^[a-z]+)')
stopwords = list()
with open('/Users/lyndon/AnacondaProjects/AnfanChen/GMO_NAS/hit_stopwords.txt', 'r', encoding='utf-8') as sw:
    for word in sw.readlines():
        stopwords.append(word.strip())

def clean_content(content):
    result = list()
    new_content = re.sub(pattern_url, '', content.strip())
    new_content = re.sub(pattern_blank, '', new_content)
    new_content = re.sub(pattern_special_char, '', new_content)
    new_content = re.sub(pattern_number, '', new_content)
    new_content = re.sub(pattern_allChar, '', new_content.lower())
    for word in jieba.lcut(new_content):
        if pattern_allChar.search(word.lower()) == None:
            word = re.sub(pattern_allChar2, '', word.lower())
            if (word.strip() not in stopwords) and (len(word.strip()) > 1):
                result.append(word.strip())
    return ' '.join(result)

training_data['new_content'] = training_data['content'].apply(clean_content)
training_data_new = training_data[['new_content', 'target']]
training_data_new.head()
```

```python
vectorizer = CountVectorizer(max_df=0.99, min_df=2)
# on training set: fit_transform
# on testing set: transform
X = vectorizer.fit_transform(training_data_new['new_content'])
print(vectorizer.get_feature_names()[10: 20])
print(X.shape)
```

```python
print('%s\n%s'%(training_data_new['target'].value_counts(), 
                training_data_new['target'].value_counts(normalize=True)))
```

```python
Y = training_data_new['target'].values
Y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, 
                                                    Y, 
                                                    test_size=0.3, 
                                                    random_state=1234)
```

```python
model = DecisionTreeClassifier(class_weight='balanced')
clf = model.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print('{:.3f}'.format(f1_score(y_test, y_pred, average="weighted")))
print('{:.3f}'.format(precision_score(y_test, y_pred, average="weighted")))
print('{:.3f}'.format(accuracy_score(y_test, y_pred))) 
print('{:.3f}'.format(recall_score(y_test, y_pred, average="weighted")))
```
